```python
def descargar_desde_url(url_csv,filename,sep = ",", delim = "\n",encoding = "utf-8"):
    import pandas as pd
    import os
    import urllib3

    #url_csv = "http://winterolympicsmedals.com/medals.csv"
    mainpath = "D:\Practica/Jupyter/"
    #filename = "Importacion de practica2.csv"
    fullpath = os.path.join(mainpath + filename)


    http = urllib3.PoolManager()
    r = http.request('GET', url_csv)
    datos = r.data.decode(encoding)

    dict = {}
    counter = 0

    filas = datos.split(delim)
    col_nombres = filas[0].split(sep)
    col_num = len(col_nombres)

    for col in col_nombres:
        dict[col] = []

    for fila in filas:
        if counter > 0:
            valor = fila.strip().split(sep)
            for i in range(col_num):
                dict[col_nombres[i]].append(valor[i])
        counter += 1

    tabla = pd.DataFrame(dict)

    tabla.to_csv(fullpath+".csv")
    tabla.to_excel(fullpath+".xlsx")
    tabla.to_json(fullpath+".json")
    print(f'Los ficheros se han guardado en {fullpath}')
    #print(tabla.head())
    return tabla
```

```python
descargar_desde_url("http://winterolympicsmedals.com/medals.csv", "prueba4")
```
